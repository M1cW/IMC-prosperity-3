```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
import pandas as pd

# Function to determine fair value using linear regression
def calculate_fair_value(timestamps, prices):
    """
    Calculate the fair value of a stock using linear regression.
    
    Parameters:
    timestamps (array): Time values
    prices (array): Mid prices
    
    Returns:
    model: Trained linear regression model
    fair_values: Predicted fair values for the given timestamps
    """
    # Reshape timestamps for sklearn
    X = np.array(timestamps).reshape(-1, 1)
    
    # Create and train the linear regression model
    model = LinearRegression()
    model.fit(X, prices)
    
    # Calculate fair values
    fair_values = model.predict(X)
    
    print(f"Fair value model - Slope: {model.coef_[0]:.8f}, Intercept: {model.intercept_:.6f}")
    return model, fair_values

# Function to predict future stock values
def predict_future_values(model_type, timestamps, prices, future_timestamps, degree=3):
    """
    Predict future stock values using either linear or polynomial regression.
    
    Parameters:
    model_type (str): 'linear' or 'polynomial'
    timestamps (array): Historical time values
    prices (array): Historical mid prices
    future_timestamps (array): Future time points to predict
    degree (int): Degree of polynomial (only for 'polynomial' model)
    
    Returns:
    model: The trained prediction model
    predictions: Predicted values for future_timestamps
    """
    # Reshape data
    X = np.array(timestamps).reshape(-1, 1)
    X_future = np.array(future_timestamps).reshape(-1, 1)
    
    if model_type.lower() == 'linear':
        model = LinearRegression()
        model.fit(X, prices)
    elif model_type.lower() == 'polynomial':
        model = make_pipeline(PolynomialFeatures(degree=degree), LinearRegression())
        model.fit(X, prices)
    else:
        raise ValueError("model_type must be either 'linear' or 'polynomial'")
    
    predictions = model.predict(X_future)
    return model, predictions

# Function to visualize the data, fair value, and predictions
def visualize_stock_analysis(timestamps, prices, fair_values, future_timestamps=None, future_values=None):
    """
    Create a visualization of stock data with fair value and optional future predictions.
    """
    plt.figure(figsize=(12, 8))
    
    # Plot historical data
    plt.scatter(timestamps, prices, color='blue', alpha=0.5, label='Mid Prices')
    
    # Plot fair value line
    plt.plot(timestamps, fair_values, color='red', label='Fair Value (Linear Regression)')
    #plt.plot(future_timestamps, fair_values, color='red', label='Fair Value (Future Prediction)')
    
    # Plot future predictions if provided
    if future_timestamps is not None and future_values is not None:
        plt.plot(future_timestamps, future_values, color='green', label='Future Prediction')
    
    plt.title('KELP Stock Analysis')
    plt.xlabel('Timestamp')
    plt.ylabel('Mid Price')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()


# df = pd.read_csv('your_kelp_data.csv')  # Uncomment and use your data source
# timestamps = df['timestamp'].values
# prices = df['mid_price'].values

# timestamps = np.linspace(0, 200000, 500)  # Example timestamps

# prices = 2025 - 0.00002 * timestamps + np.random.normal(0, 5, 500)  # Example prices
# # Example future timestamps for prediction
# future_timestamps = np.linspace(200000, 300000, 100)
# # Example future timestamps for prediction
# # Calculate fair value
# fair_model, fair_values = calculate_fair_value(timestamps, prices)
# # Predict future values using linear regression
# linear_model, future_linear_values = predict_future_values('linear', timestamps, prices, future_timestamps)
# # Predict future values using polynomial regression
# poly_model, future_poly_values = predict_future_values('polynomial', timestamps, prices, future_timestamps, degree=3)
# # Visualize the analysis
# visualize_stock_analysis(timestamps, prices, fair_values, future_timestamps, future_poly_values)
```
